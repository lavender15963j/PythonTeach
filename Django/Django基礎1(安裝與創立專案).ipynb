一、安裝
=============
使用 pip 安裝
* pip install Django

安裝完後，使用指令創立 Djnago 專案
* django-admin startproject [專案名稱]

```python
from IPython.display import Image
Image("images/P1.png")
```

cd 到專案資料夾
* cd [專案名稱]

執行專案，開啟 server
* python manage.py runserver

```python
from IPython.display import Image
Image("images/P2.png")
```

使用瀏覽器連到
* http://127.0.0.1:8000/

```python
from IPython.display import Image
Image("images/P3.png")
```
