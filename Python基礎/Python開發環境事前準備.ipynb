三、製作虛擬環境
================
這邊使用 anaconda 來製作虛擬環境。之所以要使用虛擬環境是因為每個專案所使用的第三方模組版本都有不同(如: A專案使用 django1.8，B專案使用 django2.0)，如果都使用同一個環境來跑這兩個專案會出問題(如: import 不到某某 function 之類的)，因此使用虛擬環境來解決此問題(用A環境跑A專案，B環境跑B專案，兩個環境安裝不同版本的模組)。建議每做一個專案就做一個新的環境來開發。

1. 在 cmd 輸入 conda create -n [虛擬環境名稱] python=[版本] anaconda
2. 等待完成，輸入 activate [虛擬環境名稱]，進入虛擬環境。進入環境後，cmd 的路徑會有個 ([虛擬環境名稱]) 的字樣出現(如圖)
3. 如果想要看目前電腦上創了多少虛擬環境，可以使用 conda env list 觀看
4. 要離開虛擬環境，只要輸入 deactivate 即可
